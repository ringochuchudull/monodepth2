```python
from monodepth2.usemonodepth import load_mondepthWeight, get_depthDispar, get_depthdispar_simplified, only_get_disparity

# For reading image as numpy
import cv2 
# For reading image as Pillow
from PIL import Image 
# For visualising image in Matplotlib
import matplotlib.pyplot as plt

from matplotlib import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 12 ,8
```

```python
SampleImg = './sample_images/rsz_wilkins_building_1_ucl_london_-_diliff.jpg'

pilimg = Image.open(SampleImg) 
cv2img = cv2.imread(SampleImg) 

fig, ax = plt.subplots(1,2)
ax[0].imshow(pilimg)
ax[1].imshow(cv2img)
plt.axis('off')
```

```python
# First Load Weights
encoder, depth_decoder, monoDepth_config, targetDevice = load_mondepthWeight(model_name='mono+stereo_640x192')
```

```python
# Return -> Disparity Map, Depth Map, Colorimg, Pillow Image
dis, scale_disp, colimg1, pillowim = get_depthdispar_simplified(input_image=cv2img,
                                                                model=(encoder, depth_decoder),
                                                                modelConfig=monoDepth_config,
                                                                device=targetDevice, 
                                                                channel_last=True,
                                                                isPil=False)

fig, ax = plt.subplots(1,2)
ax[0].imshow(dis)
ax[1].imshow(pillowim)
plt.axis('off')
```

```python
dis, scale_disp, colimg1, pillowim = get_depthdispar_simplified(input_image=pilimg,
                                                                model=(encoder, depth_decoder),
                                                                modelConfig=monoDepth_config,
                                                                device=targetDevice, 
                                                                channel_last=True,
                                                                isPil=True)

fig, ax = plt.subplots(1,2)
ax[0].imshow(dis)
ax[1].imshow(pillowim)
plt.axis('off')
```

```python
dis = only_get_disparity(input_image=pilimg,
                         model=(encoder, depth_decoder),
                         modelConfig=monoDepth_config,
                         device=targetDevice, 
                         channel_last=True,
                         isPil=True)
plt.imshow(dis)
```

```python
dis = only_get_disparity(input_image=cv2img,
                         model=(encoder, depth_decoder),
                         modelConfig=monoDepth_config,
                         device=targetDevice, 
                         channel_last=True,
                         isPil=False)
plt.imshow(dis)
```
